```python
import pandas as pd
import random

# Define the number of records
num_records = 100000

# Create empty lists to store data
tray_sensor = []
imaging_area_sensor = []
fusing_area_sensor = []
output_bin_sensor = []
stapler_stack_sensor = []
duplexer_sensor = []
resolved_by = [0] * num_records
resolution_count = [0] * num_records
complexity = []
suggestion = []

# Generate random data
for _ in range(num_records):
    ts = random.choice([0, 1])
    ias = random.choice([0, 1]) if ts == 1 else 0
    fas = random.choice([0, 1]) if ias == 1 else 0
    obs = random.choice([0, 1]) if fas == 1 else 0
    sss = random.choice([0, 1])
    ds = random.choice([0, 1])

    # Determine complexity based on the path the paper took
    if ts ==  1 and ias == 1 and fas == 1 and obs == 1 and ds == 1:
      comp = 3
    elif ts ==  1 and ias == 1 and fas == 1 and sss == 1 and ds == 1:
      comp = 3
    elif ts ==  1 and ias == 1 and fas == 1 and obs == 0 and sss == 0 and ds == 0:
      comp = 3
    elif ts ==  1 and obs == 1 and ds == 1 and ias == 0 and fas == 0 and sss == 0:
      comp = 2
    elif ts ==  1 and ds == 1 and obs == 1 and ias == 0 and fas == 0 and sss == 0:
      comp = 2
    elif ts ==  1 and ds == 0 and obs == 0 and ias == 0 and fas == 0 and sss == 0:
      comp = 1
    elif ts == 0 and ds == 0 and obs == 1 and ias == 0 and fas == 0 and sss == 0:
      comp = 1
    elif ts == 0 and ds == 1 and obs == 0 and ias == 0 and fas == 0 and sss == 0:
      comp = 2
    elif ts ==  0 and ds == 0 and obs == 0 and ias == 1 and fas == 0 and sss == 0:
      comp = 3
    elif ts == 0 and ds == 0 and obs == 0 and ias == 0 and fas == 1 and sss == 0:
      comp = 2
    elif ts ==  1 and ds == 0 and obs == 0 and ias == 0 and fas == 1 and sss == 0:
      comp = 2
    else:
      comp = 2



    # Determine suggestion based on complexity
    sugg = 0 if comp == 1 else 1

    # Append data to lists
    tray_sensor.append(ts)
    imaging_area_sensor.append(ias)
    fusing_area_sensor.append(fas)
    output_bin_sensor.append(obs)
    stapler_stack_sensor.append(sss)
    duplexer_sensor.append(ds)
    complexity.append(comp)
    suggestion.append(sugg)

# Create DataFrame
data = {
    'tray_sensor': tray_sensor,
    'imaging_area_sensor': imaging_area_sensor,
    'fusing_area_sensor': fusing_area_sensor,
    'output_bin_sensor': output_bin_sensor,
    'stapler_stack_sensor': stapler_stack_sensor,
    'duplexer_sensor': duplexer_sensor,
    'resolved_by': resolved_by,
    'resolution_count': resolution_count,
    'complexity': complexity,
    'suggestion': suggestion
}

df = pd.DataFrame(data)

#print(data)

# Save to CSV
#df.to_clipboard();
df.to_csv('paper_jam_data.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report

# Load the dataset
df = pd.read_csv('paper_jam_data.csv')

df.head()

# Define features and targets
features = ['tray_sensor', 'imaging_area_sensor', 'fusing_area_sensor', 'output_bin_sensor', 'stapler_stack_sensor', 'duplexer_sensor', 'resolved_by', 'resolution_count']
target_complexity = 'complexity'
target_suggestion = 'suggestion'

X = df[features]
y_complexity = df[target_complexity]
y_suggestion = df[target_suggestion]

# Split the data into training and testing sets
X_train, X_test, y_train_complexity, y_test_complexity, y_train_suggestion, y_test_suggestion = train_test_split(
    X, y_complexity, y_suggestion, test_size=0.2, random_state=42)

# Initialize and train the models
model_complexity = SGDClassifier(random_state=42)
model_suggestion = SGDClassifier(random_state=42)

model_complexity.fit(X_train, y_train_complexity)
model_suggestion.fit(X_train, y_train_suggestion)

# Make predictions and evaluate the initial models
y_pred_complexity = model_complexity.predict(X_test)
y_pred_suggestion = model_suggestion.predict(X_test)

print("Complexity Classification Report:")
print(classification_report(y_test_complexity, y_pred_complexity))

print("Suggestion Classification Report:")
print(classification_report(y_test_suggestion, y_pred_suggestion))
```
